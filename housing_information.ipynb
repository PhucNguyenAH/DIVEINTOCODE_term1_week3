- Check the percentage of missing values.

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (total/df.isnull().count()).sort_values(ascending=False)
missing = pd.concat([total, percent],axis=1,keys=["Total","percent"])
missing
```

- Delete features (columns) that have 5 or more missing values.

```python
data = df.drop(missing[missing['Total'] > 5].index,1)
data.isnull().sum().max()
```

```python
data.isnull().sum().sort_values()
```

```python
data.loc[data['Electrical'].isnull()]
```

```python
data = data.drop(data.loc[data['Electrical'].isnull()].index)
data.isnull().sum().max()
```

## Problem 5

- Skewness is a measure of symmetry, or more precisely, the lack of symmetry. A distribution, or data set, is symmetric if it looks the same to the left and right of the center point.

- Kurtosis is a measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution. That is, data sets with high kurtosis tend to have heavy tails, or outliers. Data sets with low kurtosis tend to have light tails, or lack of outliers. A uniform distribution would be the extreme case.

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
for col in data[['SalePrice']]:
    print('{:15}'.format(col), 
          'Skewness: {:05.2f}'.format(data[col].skew()) , 
          '   ' ,
          'Kurtosis: {:06.2f}'.format(data[col].kurtosis())
         )
```

```python
sns.set_theme(); np.random.seed(0)
sns.displot(data['SalePrice'], kde=False, bins=10)
```

```python
sns.set_theme(); np.random.seed(0)
target_var = data['SalePrice']
sns.displot(target_var)
plt.show()
```

```python
#log transformation

log_data = np.log(data['SalePrice'])
sns.set_theme(); np.random.seed(0)
sns.displot(log_data);

# skewness and kurtosis
print("Skewness: %f" % log_data.skew())
print("Kurtosis: %f" % log_data.kurt())
```

- Create a heat map of the correlation coefficient matrix

```python
data_corr = data.corr()
plt.figure(figsize=(16, 13))
plt.title('Correlatiom matrix')
sns.heatmap(data_corr, vmax=1, annot=False, cmap='BrBG')
plt.show()
```

- Select 10 features that have a high correlation with the target variable

```python
corr_target = data_corr[['SalePrice']].sort_values(by=['SalePrice'], ascending=False)
corr_target[1:11]
```

- Create a correlation coefficient matrix heat map for these.

```python
target_corr_matrix = data[corr_target[1:11].index]
display(target_corr_matrix)
```

```python
plt.subplots(figsize=(18,15))
plt.title('Correlatiom matrix')
sns.heatmap(target_corr_matrix.corr(), vmax=1, annot=True, cmap="YlGnBu")
plt.show()
```
